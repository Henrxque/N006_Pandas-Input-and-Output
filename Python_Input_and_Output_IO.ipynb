## Importando os Dados

```python
import pandas as pd
```

```python
#para ler JSON
nomes_f = pd.read_json('https://guilhermeonrails.github.io/nomes_ibge/nomes-f.json')
nomes_f
```

```python
nomes_m = pd.read_json('https://guilhermeonrails.github.io/nomes_ibge/nomes-m.json')
nomes_m
```

```python
#sortear nome
nomes_f.sample(5)
```

```python
#quantos nomes nÃ³s temos em cada um dos arquivos
print('Quantidade de nomes: '+ str(len(nomes_f) + len(nomes_m)))
```

```python
#unificar as listas
frames = [nomes_f, nomes_m]
frames
```

```python
#criei uma lista acima juntando os nomes_f e nomes_m, em seguida concatenei a variÃ¡vel frames que armazenou essa lista. 
#O campo que concatenei foi apenas o campo ['nome']
#to frame eu transformo a lista em um novo Dataframe
nomes = pd.concat(frames)['nome'].to_frame() 
nomes.sample(5)
```

## Como incluir IDs (Randomizar)


```python
import numpy as np
np.random.seed(123)
```

```python
total_alunos = len(nomes)
```

```python
total_alunos
```

```python
nomes["id_aluno"] = np.random.permutation(total_alunos) + 1
```

```python
nomes.sample(3)
```

```python
dominios = ['@dominiodoemail.com.br', '@servicodoemail.com']
nomes['dominio'] = np.random.choice(dominios, total_alunos)
```

```python
nomes.sample(5)
```

```python
# como montar o email: chamei o nomes['email'] que seria o nome da coluna, chamei o nomes meu df .nome(minha coluna) .str(converti em string) .cat(catch pegar) (nomes.dominio)(concatenar os dois df) #str.lower(Para deixar tudo em minÃºsculo)
nomes['email'] = nomes.nome.str.cat(nomes.dominio).str.lower()
```

```python
nomes.sample(5)
```
