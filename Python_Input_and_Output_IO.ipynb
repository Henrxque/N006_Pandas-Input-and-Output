## Importando os Dados

```python
import pandas as pd
```

```python
#unificar as listas
frames = [nomes_f, nomes_m]
frames
```

```python
#criei uma lista acima juntando os nomes_f e nomes_m, em seguida concatenei a variÃ¡vel frames que armazenou essa lista. 
#O campo que concatenei foi apenas o campo ['nome']
#to frame eu transformo a lista em um novo Dataframe
nomes = pd.concat(frames)['nome'].to_frame() 
nomes.sample(5)
```

## Como incluir IDs (Randomizar)


```python
import numpy as np
np.random.seed(123)
```

```python
total_alunos = len(nomes)
```

```python
total_alunos
```

```python
nomes["id_aluno"] = np.random.permutation(total_alunos) + 1
```

```python
nomes.sample(3)
```

```python
dominios = ['@dominiodoemail.com.br', '@servicodoemail.com']
nomes['dominio'] = np.random.choice(dominios, total_alunos)
```

```python
nomes.sample(5)
```

```python
# como montar o email: chamei o nomes['email'] que seria o nome da coluna, chamei o nomes meu df .nome(minha coluna) .str(converti em string) .cat(catch pegar) (nomes.dominio)(concatenar os dois df) #str.lower(Para deixar tudo em minÃºsculo)
nomes['email'] = nomes.nome.str.cat(nomes.dominio).str.lower()
```

```python
nomes.sample(5)
```

## Selecionando Cursos

```python
nomes.sample(5)
```

```python
#preciso saber a qtd de matrÃ­culas de cada aluno
todas_matriculas = []
#um jeito pra calcular de maneira randomica os 20 cursos
x = np.random.rand(20)
prob = x / sum(x)
```

```python
todas_matriculas = []
x = np.random.rand(6)
prob = x / sum(x)
```

```python
for index, row in nomes.iterrows():
  id = row.id_aluno
  matriculas = row.matriculas
  for i in range(matriculas):
    mat = [id, np.random.choice(cursos.index, p = prob)]
    todas_matriculas.append(mat)



matriculas = pd.DataFrame(todas_matriculas, columns = ['id_aluno', 'id_curso'])
matriculas.head(5)
```

```python
#quantos alunos tem inscritos em cada curso?
matriculas.groupby('id_curso').count().join(cursos['Nome_do_Curso']).rename(columns={'id_aluno':'quantidade_de_alunos'})
```

```python
matriculas_por_curso = matriculas.groupby('id_curso').count().join(cursos['Nome_do_Curso']).rename(columns={'id_aluno':'quantidade_de_alunos'})
matriculas_por_curso.head()
```
